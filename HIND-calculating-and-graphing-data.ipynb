Bollywood films are trending to become more western and “Hollywood” esque, in terms of clothing, 
cinematography, writing, and production. Is the style of the films also changing? Are Bollywood films 
slowly phasing out their cornerstone of music and choreography? And becoming more dialogue focused like most 
Hollywood films? Or has the trend been exactly the same or going up? 

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
```

```python
movies_df = pd.read_csv('movie.csv')

movies_df.head()
```

```python
figure(num=None, figsize=(10, 8), dpi=80, facecolor='w', edgecolor='y')
movies_df['Year Released'].value_counts()[:30].plot(kind='barh', color='gold', edgecolor='white')
plt.title('Top 30 years sampled')
plt.ylabel('Year')
plt.xlabel('Number of films from that year')
year = ['1960s', '1970s', '1980s', '1990s', '2000s', '2010s']
```

```python
movies_in_2010s = movies_df[(movies_df['Year Released'] >= 2010) & (movies_df['Year Released'] <= 2019)]
movies_in_2000s = movies_df[(movies_df['Year Released'] >= 2000) & (movies_df['Year Released'] <= 2009)]
movies_in_1990s = movies_df[(movies_df['Year Released'] >= 1990) & (movies_df['Year Released'] <= 1999)]
movies_in_1980s = movies_df[(movies_df['Year Released'] >= 1980) & (movies_df['Year Released'] <= 1989)]
movies_in_1970s = movies_df[(movies_df['Year Released'] >= 1970) & (movies_df['Year Released'] <= 1979)]
movies_in_1960s = movies_df[(movies_df['Year Released'] >= 1960) & (movies_df['Year Released'] <= 1969)]

# movies_in_2000s
```

```python
#Finding average runtime of movies
avg_film_runtime_2010s = movies_in_2010s['Runtime of Film'].mean()
avg_film_runtime_2000s = movies_in_2000s['Runtime of Film'].mean()
avg_film_runtime_1990s = movies_in_1990s['Runtime of Film'].mean()
avg_film_runtime_1980s = movies_in_1980s['Runtime of Film'].mean()
avg_film_runtime_1970s = movies_in_1970s['Runtime of Film'].mean()
avg_film_runtime_1960s = movies_in_1960s['Runtime of Film'].mean()

avg_runtime = [avg_film_runtime_1960s, avg_film_runtime_1970s, avg_film_runtime_1980s, avg_film_runtime_1990s, 
               avg_film_runtime_2000s, avg_film_runtime_2010s]

figure(num=None, figsize=(10, 5), dpi=80, facecolor='w', edgecolor='y')
plt.rc('lines', linewidth=4, color='g')
plt.plot(year, avg_runtime, color='springgreen')
plt.title('Average Film Runtime')
plt.ylabel('Film Runtime')
plt.xlabel('Decades')
plt.show()
```

```python
#Finding average song runtime of movies
avg_song_runtime_2010s = movies_in_2010s['Runtime of Song'].mean()
avg_song_runtime_2000s = movies_in_2000s['Runtime of Song'].mean()
avg_song_runtime_1990s = movies_in_1990s['Runtime of Song'].mean()
avg_song_runtime_1980s = movies_in_1980s['Runtime of Song'].mean()
avg_song_runtime_1970s = movies_in_1970s['Runtime of Song'].mean()
avg_song_runtime_1960s = movies_in_1960s['Runtime of Song'].mean()

avg_song_runtime = [avg_song_runtime_1960s, avg_song_runtime_1970s, avg_song_runtime_1980s, avg_song_runtime_1990s,
                   avg_song_runtime_2000s, avg_song_runtime_2010s]

figure(num=None, figsize=(10, 5), dpi=80, facecolor='w', edgecolor='y')
plt.rc('lines', linewidth=4, color='g')
plt.plot(year, avg_song_runtime, color='deepskyblue')
plt.title('Average Song Runtime')
plt.ylabel('Film Runtime')
plt.xlabel('Decades')
plt.show()
```

```python
#Finding the ratio of songs to films
ratio2010s = avg_song_runtime_2010s/avg_film_runtime_2010s
ratio2000s = avg_song_runtime_2000s/avg_film_runtime_2000s
ratio1990s = avg_song_runtime_1990s/avg_film_runtime_1990s
ratio1980s = avg_song_runtime_1980s/avg_film_runtime_1980s
ratio1970s = avg_song_runtime_1970s/avg_film_runtime_1970s
ratio1960s = avg_song_runtime_1960s/avg_film_runtime_1960s


ratios = [ratio1960s, ratio1970s, ratio1980s, ratio1990s, ratio2000s, ratio2010s]

figure(num=None, figsize=(10, 5), dpi=80, facecolor='w', edgecolor='y')
plt.rc('lines', linewidth=4, color='g')
plt.plot(year, ratios, color='tomato')
plt.title('Song to Film Ratio Trend')
plt.ylabel('Percentage')
plt.xlabel('Decades')
```

```python
figure(num=None, figsize=(12, 10), dpi=80, facecolor='w', edgecolor='y')
plt.rc('lines', linewidth=4, color='g')
plt.plot(year, avg_runtime, color='springgreen')
plt.plot(year, avg_song_runtime, color='deepskyblue')
plt.title('Average Film & Song Runtime')
plt.ylabel('Runtime')
plt.xlabel('Decades')
plt.show()
```

```python
#Finding avergae number of songs in each film
avg_num_songs_2010s = movies_in_2010s['# of Songs'].mean()
avg_num_songs_2000s = movies_in_2000s['# of Songs'].mean()
avg_num_songs_1990s = movies_in_1990s['# of Songs'].mean()
avg_num_songs_1980s = movies_in_1980s['# of Songs'].mean()
avg_num_songs_1970s = movies_in_1970s['# of Songs'].mean()
avg_num_songs_1960s = movies_in_1960s['# of Songs'].mean()

avg_num_songs = [avg_num_songs_1960s, avg_num_songs_1970s, avg_num_songs_1980s, avg_num_songs_1990s, avg_num_songs_2000s,
                avg_num_songs_2010s]
years_list = ["1960s", "1970s", "1980s", "1990s", "2000s", "2010s"]

print(avg_num_songs)
plt.style.use('fivethirtyeight')
figure(num=None, figsize=(12, 6), dpi=100, facecolor='w', edgecolor='y')
plt.bar(years_list, avg_num_songs, color='darkorange', alpha=0.9)
plt.title('Average number of songs in a film through the decades')
plt.ylabel('Number of Songs')
plt.xlabel('Decades')
plt.show()
```

```python
#Finding average ratio of songs through the decades

#Group Titles List
flim_len_label = ['<=90', '91-105', '106-120', '121-135', '136-150', '151-165', '166-180', '181-195', '196-210', 
              '211-225', '226-240', '241<=']

#Group null -- Films under or at 90 mins
def movie_group0(movie_group):
    return movie_group[movie_group['Runtime of Film'] <= 90] 

#Group A -- Films between 90-105 mins
def movie_groupA(movie_group):
    return movie_group[(movie_group['Runtime of Film'] > 90) & (movie_group['Runtime of Film'] <= 105)]

#Group B -- Films between 106-120 mins
def movie_groupB(movie_group):
    return movie_group[(movie_group['Runtime of Film'] > 105) & (movie_group['Runtime of Film'] <= 120)]

#Group C -- Films between 121-135 mins
def movie_groupC(movie_group):
    return movie_group[(movie_group['Runtime of Film'] > 120) & (movie_group['Runtime of Film'] <= 135)]

#Group D -- Films between 136-150 mins
def movie_groupD(movie_group):
    return movie_group[(movie_group['Runtime of Film'] > 135) & (movie_group['Runtime of Film'] <= 150)]

#Group E -- Films between 151-165 mins
def movie_groupE(movie_group):
    return movie_group[(movie_group['Runtime of Film'] > 150) & (movie_group['Runtime of Film'] <= 165)]

#Group F -- Films between 166-180 mins
def movie_groupF(movie_group):
    return movie_group[(movie_group['Runtime of Film'] > 165) & (movie_group['Runtime of Film'] <= 180)]

#Group G -- Films between 181-195 mins
def movie_groupG(movie_group):
    return movie_group[(movie_group['Runtime of Film'] > 180) & (movie_group['Runtime of Film'] <= 195)]

#Group H -- Films between 196-210 mins
def movie_groupH(movie_group):
    return movie_group[(movie_group['Runtime of Film'] > 195) & (movie_group['Runtime of Film'] <= 210)]

#Group I -- Films between 211-225 mins
def movie_groupI(movie_group):
    return movie_group[(movie_group['Runtime of Film'] > 210) & (movie_group['Runtime of Film'] <= 225)]

#Group J -- Films between 226-240 mins
def movie_groupJ(movie_group):
    return movie_group[(movie_group['Runtime of Film'] > 210) & (movie_group['Runtime of Film'] <= 240)]

#Group K -- Films beyond 241 mins
def movie_groupK(movie_group):
    return movie_group[movie_group['Runtime of Film'] > 240] 
```

```python
#2010s expanded data set

#Creating a list calling the diffferent runtime functions
rof_2010s_list = [movie_group0(movies_in_2010s), movie_groupA(movies_in_2010s), movie_groupB(movies_in_2010s),
                 movie_groupC(movies_in_2010s), movie_groupD(movies_in_2010s), movie_groupE(movies_in_2010s), 
                 movie_groupF(movies_in_2010s), movie_groupG(movies_in_2010s), movie_groupH(movies_in_2010s),
                 movie_groupI(movies_in_2010s), movie_groupJ(movies_in_2010s), movie_groupK(movies_in_2010s)]

#Creating lists to find song len, movie len, and num songs averages
song_len_avg_2010s = []
movie_len_avg_2010s = []
num_songs_avg_2010s = []
avg_film_rating_2010s = []

#Appending the averages of each category to its respective list
for movie_df in rof_2010s_list:
    song_len_avg_2010s.append(movie_df['Runtime of Song'].mean()) # change runtime to song runtime
    movie_len_avg_2010s.append(movie_df['Runtime of Film'].mean())
    num_songs_avg_2010s.append(movie_df['# of Songs'].mean())
#     avg_film_rating_2010s.append(movie_df['Film Rating'].mean())
    
#Testing and printing to see if the data works    
print("Song length Avg:", song_len_avg_2010s)
print("Number of songs avg:", num_songs_avg_2010s)

#Graph Results
# figure(num=None, figsize=(10, 5), dpi=80, facecolor='w', edgecolor='y')
# plt.bar(flim_len_label, song_len_avg_2010s)
# plt.show()
```
